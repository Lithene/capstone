#### Instantition of facebook scraper

```python
#instantiate the Facebook_scraper class
#page_name = "metaai"
posts_count = 100
browser = "firefox"
#proxy = "IP:PORT" #if proxy requires authentication then user:password@IP:PORT
timeout = 60 #600 seconds
headless = True
#meta_ai = Facebook_scraper(page_name, posts_count, browser, proxy=proxy, timeout=timeout, headless=headless)
```

```python
def extract_json(page_name):
    #all_data = []

    #call scrap_to_json(filename,directory) method
    #for page_name in page_list:
    page_to_scrap = Facebook_scraper(page_name, posts_count, browser, timeout=timeout, headless=headless) #proxy=proxy
    json_data = page_to_scrap.scrap_to_json()
    #all_data.append(json_data)
    
    #all_json = ','.join(all_data)
    df = pd.read_json(StringIO(json_data), orient='columns') #, lines=True, chunksize=1
    df = df.transpose()
    
    return df
```

```python
df1 = extract_json("NicholasGohOrganisation")
df2 = extract_json("QualityLifeWithQiLe")
df3 = extract_json("AinWallofTrust")
df4 = extract_json("JayceeOngFC")
df5 = extract_json("darrenfinancialconsultant")
```

```python
output = df1.append([df2, df3, df4, df5])
```
